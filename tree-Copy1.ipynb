```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.datasets import load_boston
# boston為連續的值
```

```python
import pandas as pd
boston = load_boston()
df = pd.DataFrame(boston["data"], columns=boston["feature_names"])
df["target"] = boston["target"] 
df
```

```python
# 切開來 -> 為之後步驟做準備
from sklearn.model_selection import train_test_split
# (90%題目，10%題目，90%答案，10%答案)
# iris 可顯示訓練好的結果
# train為90%，test為10%
x_train, x_test, y_train, y_test = train_test_split(boston["data"], boston["target"], test_size=0.1)
```

```python
# (90%) (訓練) 算出方程式 ---> fit
reg = DecisionTreeRegressor(max_depth=4)
# max_depth 後寫你要幾層(如果較小2或3即可)
# 當你建模時clf就已經必成件磨好的檔案
reg.fit(x_train, y_train)
```

```python
import graphviz
from sklearn.tree import export_graphviz
r = export_graphviz(reg, feature_names=boston["feature_names"], filled=True)
# 寫feature_names只是為了增加內容
graphviz.Source(r) # 畫圖
# 分到只有一類為止(把不符合趨勢(最後兩個)的東西刪除)
```

```python
# 因為訓練模型取90%所以做出來都不太一樣
# 決策樹:解釋性強
```

```python
from sklearn.metrics import accuracy_score
# (10%) (測試) 驗證方程式好不好
pre = reg.predict(x_test)
print("預測:", list(pre))
print("正確:", list(y_test))
# 用預測跟正確來做比較
```

```python
from sklearn.metrics import r2_score
print("r2_score:", r2_score(y_test, pre))
```
